```python
import numpy as np
import string
import pymorphy2
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
def lemmatization(words):
    morph = pymorphy2.MorphAnalyzer(lang='ru')
    lemmas = {}

    for item in words:
        print(item)
        lemma = morph.parse(item)[0].normal_form
        if lemma in lemmas:
            if item not in lemmas[lemma]:
                lemmas[lemma].append(item)
        else:
            lemmas[lemma] = [item]

    return lemmas
```

```python
def tokenization(text):
    result = []
    r = re.compile("^[а-яА-ЯёЁ]+$") 

    for line in text:
        
        line = line.lower()
        line = re.sub(r'[^\w\s]','', line)
        line = [t for t in line.split() if t not in stopwords.words('russian')]
        temp = [word for word in filter(r.match, line)]
        result += temp

    return result
```

```python
file_number = 1

while file_number<=165:
    with open('files/file_{0}.txt'.format(file_number), encoding='utf-8') as file:
        text = [row.strip() for row in file]
        tokens = tokenization(text)
        tokens = set(tokens)
        tokens = list(tokens)
        with open('files/words.txt'.format(file_number), 'a', encoding='utf-8') as file:
            for item in tokens:
                file.write("%s\n" % item)
    file_number = file_number + 1
```

```python
f = open('files/words.txt', encoding='utf-8')
words = []
for line in f:
    words.append(line.strip())
```

```python
lemmas = lemmatization(words)

with open('files/lemmas.txt','a') as file:
    for key,val in lemmas.items():
        file.write('\n{} '.format(key))
        for item in val:
            file.write('{} '.format(item))
```
